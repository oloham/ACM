```python
#checking unique values again
unique_values = pts_1.values.ravel()
unique_values = pd.Series(unique_values).unique()
unique_values
```

```python
#creating a column with sums of rows and sorting rows accordingly in ascending order 

pts_1_copy = pts_1.copy()
pts_1_copy.loc[:, 'sum_rows'] = pts_1_copy.sum(axis='columns')
pts_1_copy = pts_1_copy.sort_values(by=['sum_rows']).reset_index(drop=True)
pts_1 = pts_1_copy
```

```python
#replacing 2,3,4 values with zeros; checking values 

pts_1_1=pts_1.replace({2:0,3:0,4:0})
unique_values = pts_1_1.values.ravel()
unique_values = pd.Series(unique_values).unique()
unique_values
```

```python
#replacing 1,3,4 values with zeros, and 2 with 1; checking values 

pts_1_2=pts_1.replace({1:0,2:1,3:0,4:0})
unique_values = pts_1_2.values.ravel()
unique_values = pd.Series(unique_values).unique()
unique_values
```

```python
#replacing values 1,2,4 with zeros, and 3 with one; checking values 

pts_1_3=pts_1.replace({1:0,2:0,3:1,4:0})
unique_values = pts_1_3.values.ravel()
unique_values = pd.Series(unique_values).unique()
unique_values
```

```python
#replacing values 1,2,3 with zeros, and 4 with one; checking values 

pts_1_4=pts_1.replace({1:0,2:0,3:0,4:1})
unique_values = pts_1_4.values.ravel()
unique_values = pd.Series(unique_values).unique()
unique_values
```

```python
#grouping rows with identical 'sum_rows' by mean values, result being probability to answer question in certain way for specified score 

pts_1_1=pts_1_1.groupby('sum_rows', as_index=False).mean()
column_to_move = pts_1_1.pop('sum_rows')
pts_1_1.insert(len(pts_1_1.columns), 'sum_rows', column_to_move)

pts_1_2=pts_1_2.groupby('sum_rows', as_index=False).mean()
column_to_move = pts_1_2.pop('sum_rows')
pts_1_2.insert(len(pts_1_2.columns), 'sum_rows', column_to_move)

pts_1_3=pts_1_3.groupby('sum_rows', as_index=False).mean()
column_to_move = pts_1_3.pop('sum_rows')
pts_1_3.insert(len(pts_1_3.columns), 'sum_rows', column_to_move)

pts_1_4=pts_1_4.groupby('sum_rows', as_index=False).mean()
column_to_move = pts_1_4.pop('sum_rows')
pts_1_4.insert(len(pts_1_4.columns), 'sum_rows', column_to_move)
```

```python
#rounding probability values to second decimal 

pts_1_1=round(pts_1_1,2)
pts_1_2=round(pts_1_2,2)
pts_1_3=round(pts_1_3,2)
pts_1_4=round(pts_1_4,2)
```

```python
#creating function that renders three matrices: 
#with frequency of occurence of random indices drawn from a flattened matrix, 
#with ones and zeros representing comparison between randomly drawn values , 
#with ones representing successful comparison for the value on the left 

def calculate_scores(pts):
    rows_to_lists = pts.iloc[:, :-1].values.tolist()
    flat_list = [item for sublist in rows_to_lists for item in sublist]
    acm_list = []
    freq_list = [0 for _ in flat_list]
    item_score = [0 for _ in flat_list]

    for x in range(len(flat_list)):
        indices = []
        acm_prep = []
        acm_prep.append((flat_list[x], x))

        random_num = np.random.choice(
            range(x + 1, len(flat_list)), 
            size=math.ceil((len(flat_list) - x - 1) / 2),
            replace=False)

        indices.extend(random_num)
        sorted_indices = sorted(indices)

        for index in sorted_indices:
            freq_list[index] += 1
            freq_list[x]+=1
            acm_prep.append((flat_list[index], index))
        for y in range(len(acm_prep) - 1):
            if acm_prep[0][0] < acm_prep[y + 1][0]:
                acm_list.append(1)
                item_score[acm_prep[0][1]] += 1
            else:
                acm_list.append(0)
    return freq_list, acm_list, item_score
```

```python
#calling function calculate_scores

freq_list_1_1, acm_list_1_1, item_score_1_1 = calculate_scores(pts_1_1)
freq_list_1_2, acm_list_1_2, item_score_1_2 = calculate_scores(pts_1_2)
freq_list_1_3, acm_list_1_3, item_score_1_3 = calculate_scores(pts_1_3)
freq_list_1_4, acm_list_1_4, item_score_1_4 = calculate_scores(pts_1_4)
```

```python
#transforming item_score_ndarray to matrices

df_item_score_1_1=pd.DataFrame(item_score_ndarray_1_1, columns=pts_1_1.iloc[:,:-1].columns)
df_freq_list_1_1=pd.DataFrame(freq_list_ndarray_1_1, columns=pts_1_1.iloc[:,:-1].columns)

df_item_score_1_2=pd.DataFrame(item_score_ndarray_1_2, columns=pts_1_2.iloc[:,:-1].columns)
df_freq_list_1_2=pd.DataFrame(freq_list_ndarray_1_2, columns=pts_1_2.iloc[:,:-1].columns)

df_item_score_1_3=pd.DataFrame(item_score_ndarray_1_3, columns=pts_1_3.iloc[:,:-1].columns)
df_freq_list_1_3=pd.DataFrame(freq_list_ndarray_1_3, columns=pts_1_3.iloc[:,:-1].columns)

df_item_score_1_4=pd.DataFrame(item_score_ndarray_1_4, columns=pts_1_4.iloc[:,:-1].columns)
df_freq_list_1_4=pd.DataFrame(freq_list_ndarray_1_4, columns=pts_1_4.iloc[:,:-1].columns)
```

```python
#calculating quality of questions through ACM

item_quality_1_1 = df_item_score_1_1.div(df_freq_list_1_1)
item_quality_1_2 = df_item_score_1_2.div(df_freq_list_1_2)
item_quality_1_3 = df_item_score_1_3.div(df_freq_list_1_3)
item_quality_1_4 = df_item_score_1_4.div(df_freq_list_1_4)
```
